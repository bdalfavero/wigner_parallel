```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import scipy.linalg as la
import h5py
```

```python
dir = "../data"
data_files = os.listdir(dir)
domain_data_files = [s for s in data_files if "domain_" in s and "h5" in s]
domain_h5_files = [h5py.File(dir + '/' + fn) for fn in convergence_data_files]
```

```python
walltimes = np.array([hf["total_walltime"][()] for hf in domain_h5_files])
num_points = np.array([hf["num_points"][()] for hf in domain_h5_files])
```

```python
fig, ax = plt.subplots()
ax.plot(num_points, walltimes, '.')
ax.set_xlabel("Points in domain")
ax.set_ylabel("Total walltime")
plt.savefig("../plots/domain_scaling.png")
```
